```python
import matplotlib.pyplot as plt
import numpy as np

# Test labels
tests = [
    "Test 1",
    "Test 2",
    "Test 3",
    "Test 4",
    "Test 5",
    "Test 6"
]

# MAE and RMSE values
mae = [101.7921, 75.6139, 76.8578, 68.0230, 66.0522, 66.9522]
rmse = [340.171832, 285.9800, 287.5943, 259.5497, 249.3955, 257.3482]

x = np.arange(len(tests))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width/2, mae, width, label='MAE')
bar2 = ax.bar(x + width/2, rmse, width, label='RMSE')

ax.set_xlabel('Test Variants')
ax.set_ylabel('Error Value')
ax.set_title('XGBoost Prediction Errors for Different Dataset Variants')
ax.set_xticks(x)
ax.set_xticklabels(tests, rotation=0)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with values
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
